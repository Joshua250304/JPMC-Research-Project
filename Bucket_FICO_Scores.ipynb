```python
import pandas as pd
import numpy as np

# 1. Load the data (Same as Task 3)
df = pd.read_csv('Loan_Data.csv')

# --- Start of suggested change ---
# Print DataFrame columns to identify the correct FICO score column name
print("DataFrame columns:", df.columns.tolist())
# --- End of suggested change ---

def bucket_fico_scores(df, num_buckets=5):
    """
    Groups FICO scores into buckets and calculates the probability
    of default for each bucket.
    """

    # Method A: Quantile Bucketing (Equal number of people in each bucket)
    # Replace '[CORRECT_FICO_SCORE_COLUMN_NAME]' with the actual column name from the printout above
    df['fico_bucket'] = pd.qcut(df['fico_score'], q=num_buckets, labels=False)

    # Method B: Custom Boundary Bucketing (Optional - for more control)
    # bins = [300, 580, 670, 740, 800, 850]
    # labels = ['Very Poor', 'Fair', 'Good', 'Very Good', 'Exceptional']
    # df['fico_bucket_custom'] = pd.cut(df['ficoscore'], bins=bins, labels=labels)

    # 2. Analyze the Default Rate per Bucket
    # This helps us see if the buckets actually predict risk effectively
    bucket_analysis = df.groupby('fico_bucket').agg({
        # Replace '[CORRECT_FICO_SCORE_COLUMN_NAME]' with the actual column name
        'fico_score': ['min', 'max'],
        'default': 'mean',  # This is the Probability of Default for this bucket
        'customer_id': 'count'
    }).reset_index()

    # Renaming columns for clarity
    bucket_analysis.columns = ['Bucket', 'Min_FICO', 'Max_FICO', 'Default_Rate', 'Count']

    return bucket_analysis

# 3. Dynamic Programming Approach (Conceptual logic)
# To find 'optimal' buckets, we want to minimize the variance of 'default'
# within each bucket while maximizing the difference between buckets.

# The function call will now wait for the user to replace the placeholder
# with the actual FICO score column name.
optimal_buckets = bucket_fico_scores(df)
print(optimal_buckets)
```
